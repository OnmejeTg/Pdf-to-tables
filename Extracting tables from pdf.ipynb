```python
# for data analysis
import pandas as pd
from matplotlib import pyplot
import seaborn as sns

# for extracting pdf
import camelot as cm
```

```python
input_pdf = cm.read_pdf('NGA.pdf', flavor='lattice', pages='all')
```

```python
input_pdf
```

```python
for n in input_pdf:
    print(n)
```

```python
df = input_pdf[0].df
df
```

```python
df.columns = ['Year', 'Life expectancy at birth', 'Expected years of schooling', 'Mean years of schooling', 'GNI per capita (2017 PPP$)', 'HDI value']
```

```python
df.drop([0], inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
df.drop(['index'], axis=1, inplace=True)
```

```python
df
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df['Life expectancy at birth']  = df['Life expectancy at birth'].astype(float)
```

```python
sns.barplot(x='Year', y='Life expectancy at birth', data=df);
```

```python
# to save table as csv
df.to_csv('NGA.csv')
```
